```python
import scipy.stats as stats
```

```python
# Bivariate analysis
bmi_diabetes = pd.crosstab(df['bmi_cat'], df['Diabetes']) # contingency table of BMI categories and diabetes status
chi2, p, dof, expected = stats.chi2_contingency(bmi_diabetes) # chi-square test for association between BMI and diabetes
print('Chi-square test results:\nChi-square = {}\np-value = {}\nDegrees of freedom = {}'.format(chi2, p, dof))
```

```python
# Interpretation and conclusion
if p < 0.05:
    print('There is a significant association between BMI and diabetes in the dataset.')
else:
    print('There is no significant association between BMI and diabetes in the dataset.')
```

```python
# Create a contingency table of diabetes and sex
diabetes_sex = pd.crosstab(df['Diabetes'], df['Sex'])

# Perform chi-square test of independence
chi2, p, dof, expected = stats.chi2_contingency(diabetes_sex)

# Print the results
print('Chi-square test results:\nChi-square = {}\np-value = {}\nDegrees of freedom = {}'.format(chi2, p, dof))

# Interpret the results
if p < 0.05:
    print('There is a significant association between diabetes and sex in the dataset.')
    
    # Calculate the proportions of diabetes by sex
    female_diabetes = diabetes_sex.loc[1, 0] / diabetes_sex[0].sum()
    male_diabetes = diabetes_sex.loc[1, 1] / diabetes_sex[1].sum()
    
    # Compare the proportions of diabetes by sex
    if female_diabetes > male_diabetes:
        print('Females have a higher chance of getting diabetes compared to males in the dataset.')
    else:
        print('Males have a higher chance of getting diabetes compared to females in the dataset.')
else:
    print('There is no significant association between diabetes and sex in the dataset.')
```

```python
# Reading csv after adding interaction variable Fruits & veggies
df1 = pd.read_csv("/content/diabetes_data1.csv")
 
# Printing top 5 rows
df1.head()
```

```python
# counting total number of missing numbers
count_nan = df1.isna().sum().sum()
print('Total missing numbers are '+str(count_nan))
print(df1.isna().sum())
```

```python
import statsmodels.api as sm
model = sm.Logit(df1['Diabetes'], sm.add_constant(df1[['BMI', 'Sex', 'Age', 'PhysActivity', 'Fruits', 'Veggies', 'Fruits & Veggies', 'HighBP']])) # logistic regression model
results = model.fit()
print(results.summary()) # model summary
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Predict the values of the dependent variable using the trained model
y_pred = results.predict()

# Convert the predicted probabilities to binary values using a threshold of 0.5
y_pred_binary = [1 if p >= 0.5 else 0 for p in y_pred]

# Compute the confusion matrix
cm = confusion_matrix(df1['Diabetes'], y_pred_binary)

# Compute the performance metrics
accuracy = accuracy_score(df1['Diabetes'], y_pred_binary)
precision = precision_score(df1['Diabetes'], y_pred_binary)
recall = recall_score(df1['Diabetes'], y_pred_binary)
f1 = f1_score(df1['Diabetes'], y_pred_binary)

# Print the results
print('Confusion Matrix:\n', cm)
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1)
```
